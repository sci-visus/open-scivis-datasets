```python
import matplotlib.pyplot as plt
import OpenVisus as ov
```

```python
print(help(ov.load_dataset))
dataset = ov.load_dataset('https://klacansky.com/open-scivis-datasets/tooth/tooth.idx', cache_dir='.')
print('Max resolution', dataset.max_resolution)
print('Field names', dataset.field_names)
```

```python
# read whole volume at full resolution
print(help(dataset.read))
data = dataset.read()
data.shape
```

```python
# read box in voxel space (x=[inclusive,exclusive])
data = dataset.read(x=(16,48), z=(10,50))
data.shape
```

```python
# read x slice
data = dataset.read(x=50)
plt.imshow(data);
```

```python
# read y slice
data = dataset.read(y=32)
plt.imshow(data);
```

```python
# read z slice
data = dataset.read(z=20)
plt.imshow(data);
```
